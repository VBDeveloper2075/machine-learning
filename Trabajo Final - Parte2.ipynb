Parte 1
Resolver el siguiente ejercicio de programación El empleado llamado Juan cobró 300 dólares por mes desde enero a junio, 500 dólares de julio a octubre, y 700 dólares por mes en noviembre y en diciembre.

En base al escenario propuesto, se le solicita a los estudiantes que creen un pequeño programa que calcule el sueldo promedio del empleado Juan. Además, se debe indicar sí Juan se encuentra cobrando un sueldo bajo, normal o mejor de lo normal, considerando las siguientes pautas:

a. Sueldo bajo: por debajo de 300 dólares. b. Sueldo normal: entre 300 a 900. c. Sueldo mejor de lo normal: más de 900 dólares.

Tip: se debe utilizar estructuras condicionales.

```python
# Definir los montos de sueldo por mes
sueldo_enero_junio = 300
sueldo_julio_octubre = 500
sueldo_noviembre_diciembre = 700

# Calcular el sueldo promedio
sueldo_promedio = (sueldo_enero_junio * 6 + sueldo_julio_octubre * 4 + sueldo_noviembre_diciembre * 2) / 12

# Determinar la categoría de sueldo
categoria_sueldo = ""
if sueldo_promedio < 300:
    categoria_sueldo = "Sueldo bajo"
elif sueldo_promedio <= 900:
    categoria_sueldo = "Sueldo normal"
else:
    categoria_sueldo = "Sueldo mejor de lo normal"

# Imprimir los resultados
print("El sueldo promedio de Juan es:", sueldo_promedio)
print("Categoría de sueldo:", categoria_sueldo)
```
